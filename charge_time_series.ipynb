```python
%matplotlib inline
import numpy as np
import sys

import osh5io
import osh5def
import osh5vis
import osh5utils
import osh5dir
import matplotlib.pyplot as plt
```

```python
#base_directory = '/home/agrt/osiris_diiorio/osiris/data/test' #_no_secondary'
#osh5dir.runinfo(base_directory)
#base_directory = '/scratch/agrt_root/agrt0/agrt/test_fld1'
base_directory = '/home/agrt/osiris_diiorio/osiris/data/test_coll'

```

```python
nmax = 500
ntest = 10
#ntest = 7

species = 'etarget'
field = 'charge'
charge_t = np.zeros((ntest,nmax))


for tests in range(0,ntest):
    print(tests,"of",ntest)
    (field_name,fldmap) = osh5dir.getfieldname(base_directory+str(tests+1),field,species,'')
    for filenum in range(0,nmax):
        data = osh5dir.getfile(field_name,filenum)
        charge_t[tests,filenum] = -np.sum(data)    
```

```python
time = 2.5* (np.arange(0,nmax))
for tests in range(0,ntest):
    plt.plot(time,charge_t[tests,:])
plt.legend(('NPPC=1','NPPC=2','NPPC=4','NPPC=8','NPPC=16','NPPC=32','NPPC=64','NPPC=128','NPPC=256','NPPC=512'))
plt.xlabel('$\omega_p t$')
plt.ylabel('$n_e/n_0$')
```
